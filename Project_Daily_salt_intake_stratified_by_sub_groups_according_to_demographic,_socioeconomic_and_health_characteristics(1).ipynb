```python
#open Daily salt intake stratified by sub-groups according to demographic, socioeconomic and health characteristics - Sheet1.csv as pandas dataframe
import pandas as pd
saltintake_df = pd.read_csv("Daily salt intake stratified by sub-groups according to demographic, socioeconomic and health characteristics - Sheet1.csv")
#Scan first five rows and columns
saltintake_df.head(20)
```

```python
#Preprocessing I - remove unwanted columns - unamed 1, unamed 2, unmaed 4, unamed 7 
newsaltintake_df = saltintake_df.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 7'])
#Scan columns
newsaltintake_df.head(20)
```

```python
#Preprocessing II - drop all rows with all missing values = rows 0, 12, 17, 22 and 28 
dropNa_df = newsaltintake_df.dropna(how='all')

#Scan rows
dropNa_df.head(50)
```

```python
#Preprocessing III - replace all NaN values with empty string
#import numpy library/module

import numpy as np

finalsaltintake_df = dropNa_df.replace(np.nan, '', regex=True)

finalsaltintake_df.head(50)

```

```python
#check n samples
#Sub-group age
finalsaltintake_df.count(53, 51, 68, 58)
```

```python
#Data visualisation - vertical bar chart
#Age and salt intake
df = pd.DataFrame({'Age': ['<30', '30-39', '40-49', '50-59', '>60' ], 'salt intake (mEq/L)': [10.3, 10.3, 10.8, 10.4, 9.8]})
ax = df.plot.bar(x='Age', y='salt intake (mEq/L)')
```

```python
#Data visualisation - horizontal bar chart
#Age and salt intake
df = pd.DataFrame({'Age': ['<30', '30-39', '40-49', '50-59', '>60' ], 'salt intake (mEq/L)': [10.3, 10.3, 10.8, 10.4, 9.8]})
ax = df.plot.barh(x='Age', y='salt intake (mEq/L)')
```

```python
#Data visualisation - vertical bar chart
#Race and salt intake
df = pd.DataFrame({'Race': ['White', 'Brown', 'Black'], 'salt intake (mEq/L)': [10.1, 10.4, 11.2]})
ax = df.plot.bar(x='Race', y='salt intake (mEq/L)')
```

```python
#Data visualisation - horizontal bar chart
#Race and salt intake
df = pd.DataFrame({'Race': ['White', 'Brown', 'Black'], 'salt intake (mEq/L)': [10.1, 10.4, 11.2]})
ax = df.plot.barh(x='Race', y='salt intake (mEq/L)')
```

```python
#Data visualisation - vertical bar chart
#Body mass index and salt intake
df = pd.DataFrame({'Body mass index': ['<25', '25-29.9', '>30'], 'salt intake (mEq/L)': [9.2, 10.7, 11.7]})
ax = df.plot.bar(x='Body mass index', y='salt intake (mEq/L)')
```

```python
#Horizontal bar chart

index = ['White', 'Brown', 'Black']

df = pd.DataFrame({'Body mass index': ['<25', '25-29.9', '>30'], 'salt intake (mEq/L)': [9.2, 10.7, 11.7]}, index=index)

ax = df.plot.barh()
```
